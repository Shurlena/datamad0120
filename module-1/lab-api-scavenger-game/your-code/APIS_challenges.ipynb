### Challenge 1: Fork Languages

```python
load_dotenv(".env")

def requestsGithub(endpoint):
    token = os.getenv("API_TOK")
    baseUrl = "https://api.github.com"
    url = baseUrl+endpoint
    headers = {
        "Authorization": f"token {token}"
    }
    res = requests.get(url,headers=headers)
    return res.json()

requestsGithub("/repos/ironhack-datalabs/datamad0120/forks")
```

```python
data = requestsGithub("/repos/ironhack-datalabs/datamad0120/forks")

language = ([e["language"] for e in data])

print(list(set(language)))
```

### Challenge 3: Hidden Cold Joke

##### 1. Saco el SHA:

##### 2. Saco las carpetas con sus archivos:

```python
requestsGithub("/repos/ironhack-datalabs/scavenger/git/trees/9308ccc8a4c34c5e3a991ee815222a9691c32476?recursive=1")
```

##### 3. Saco la lista con el path de las carpetas y los archivos

```python
data = requestsGithub("/repos/ironhack-datalabs/scavenger/git/trees/9308ccc8a4c34c5e3a991ee815222a9691c32476?recursive=1")

paths = [e['path'] for e in data['tree'] if ".scavengerhunt" in e['path']]
print(paths)
```

##### 4. Saco el contenido de cada archivo junto con el nombre para poder ordenarlo:

```python
contents = []

for path in paths:
    contents.append((requestsGithub(f'/repos/ironhack-datalabs/scavenger/contents/{path}')['name'] ,requestsGithub(f'/repos/ironhack-datalabs/scavenger/contents/{path}')['content']))

contents = sorted(contents)
print(contents)
```

##### 5. Una vez ordenado, me quedo solo con el contenido:

```python
ordered_contents = []

for i in contents:
    ordered_contents.append(i[1])
    
print(ordered_contents)
```

##### 6. Decodifico y saco la frase:

```python
import base64

frase = ""

for e in ordered_contents:
    a = base64.b64decode(e)
    a = a.decode("utf-8")
    frase += a.replace("\n", " ")
    
print(frase)
```
